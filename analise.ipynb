```python
import pandas as pd
import numpy as np

df = pd.read_csv('dados-spfc-limpo.csv')
print(df.head(1))
print(df.columns)
```

```python
# Vitórias, Empates e Derrotas

vitorias = (df['resultado'] == 'V').sum()
derrotas = (df['resultado'] == 'D').sum()
empates = (df['resultado'] == 'E').sum()

print(f'O total de vitorias é : {vitorias}')
print(f'O total de derrotas é : {derrotas}')
print(f'O total de empates é : {empates}')
```

```python
# Saldo de gols

saldo = df['saldo'].sum()

print(f'O saldo de gols é: {saldo}')
```

```python
# aproveitamento

pontos_conquistados = (vitorias * 3) + (empates*1)
pontos_disputados = (df['rodada'].value_counts()) * 3
aproveitamento = (pontos_conquistados / pontos_disputados) * 100
print(aproveitamento)
```

```python
# maiores vitimas

df_jogos = df.copy()

df_jogos['adversario'] = np.where(df_jogos['casa'] != 'São Paulo', df_jogos['casa'], df_jogos['visitante'])

ranking = df_jogos.groupby('adversario')['resultado'].value_counts().unstack(fill_value=0)
ranking = ranking.reset_index()
ranking['total_jogos'] = ranking['V'] + ranking['E'] + ranking['D']
ranking = ranking.sort_values(by='V', ascending=False)

```
